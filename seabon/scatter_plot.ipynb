```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic_df = sns.load_dataset('titanic')
titanic_df
```

```python
sns.scatterplot(x = "age",y = 'fare',data = titanic_df)
```

```python
sns.scatterplot(x = "age",y = 'fare',data = titanic_df,hue = 'sex')
```

```python
sns.scatterplot(x = "age",y = 'fare',data = titanic_df,hue = 'sex',
                style = 'who')
```

```python
sns.scatterplot(x = "age",y = 'fare',data = titanic_df,
                 hue = 'sex',style = 'who',size = 'who') 
```

```python
sns.scatterplot(x = "age",y = 'fare',data = titanic_df,
                 hue = 'sex',style = 'who',size = 'who',sizes = (100,400)) 
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x = "age",y = 'fare',data = titanic_df,
                 hue = 'sex',style = 'who',size = 'who',sizes = (100,400)) 
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x = "who",y = 'fare',data = titanic_df,
                 hue = 'alive',style = 'who',size = 'who',sizes = (100,400)) 
```

```python
plt.figure(figsize = (16,9))
sns.scatterplot(x = 'age',y = 'fare',data = titanic_df)
sns.lineplot(x = 'age',y='fare',data = titanic_df)
sns.barplot(x = 'age',y = 'fare',data = titanic_df)
```
